```python
import requests
import time
from bs4 import BeautifulSoup

# 列出10支股票30天內的股價

# 先算出30日曆天前的日期
day30 = time.strftime("%y/%m/%d",
              time.localtime(time.mktime(time.localtime(time.time())) - 60 * 60 * 24 * 30)).lstrip('0')
date=day30.split("/")

# 用來跟股價資料比對的格式
day30str = str(int(date[0])+89) + '/' + date[1] + '/' +date[2]
# 放進URL的日期格式
urldate = str(int(date[0])+2000)+date[1]+date[2]

#print(day30str, urldate)

selectstocks = {'2330', '2303', '2454', '0050', '00692', '00892', '3008', '2006', '2002', '00895'}

#url:　https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=html&stockNo=2330&date=20220526

for stock in selectstocks:
    print(stock)
    data = []
    # 先找上個月的股票資料
    url = 'https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=html&stockNo=' + stock + '&date=' + urldate
    resp = requests.get(url)
    if resp.status_code == requests.codes.ok:
        soup = BeautifulSoup(resp.text, 'html5lib')
        tables = soup.find('table').find_all('tr')
        tr_count = 1
        for tr in tables:
            if (tr_count > 2):
                datas = tr.find_all('td')
                count = 1
                for td in datas:
                    if (count == 1):
                        date = td.text
                    if (count == 7):
                        price = td.text
                    count += 1

                if (date >= day30str):
                    data.append({
                        'stock': stock,
                        'date': date,
                        'price': price
                    })
            tr_count += 1
    #再找這個月的資料
    url = 'https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=html&stockNo=' + stock
    resp = requests.get(url)
    if resp.status_code == requests.codes.ok:
        soup = BeautifulSoup(resp.text, 'html5lib')
        tables = soup.find('table').find_all('tr')
        tr_count = 1
        for tr in tables:
            if (tr_count > 2):
                datas = tr.find_all('td')
                count = 1
                for td in datas:
                    if (count == 1):
                        date = td.text
                    if (count == 7):
                        price = td.text
                    count += 1

                data.append({
                    'stock': stock,
                    'date': date,
                    'price': price
                })
            tr_count += 1

    if data:
        for item in data:
            print("{:>10}".format(item['date']), end=' ')
            print("{:>8}".format(item['price']), end=' ')
    print()


    # data.append(str(data))
        # print(datas[0].text, datas[1].text, datas[2].text, datas[3].text, datas[4].text, datas[5].text, datas[6].text, datas[7].text, datas[8].text)
```
