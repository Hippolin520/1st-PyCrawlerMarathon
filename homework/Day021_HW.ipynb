## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import date
import time
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
def list_all_news(soup):
    print("--- Total news:", len(soup.find(class_="part_list_2").find_all('h3')))
    for d in soup.find(class_="part_list_2").find_all('h3'):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
def get_last_update(soup):
    v_no = 0
    for s in soup.find(class_="part_list_2").find_all('h3'):
        v_date, v_time = s.find(class_="date").text.split(" ")
        v_subject = s.find_all('a')[-1].text
        v_no +=1
    return (v_no, v_date, v_time, v_subject)
```

```python
# 每個兩秒鐘自動往下滑
for r in range(50):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight),1000;")
```

```python
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
last_no, last_date, last_time, last_subject = get_last_update(soup)
print(last_no, last_date, last_time)
```

```python
today = time.strftime("%Y/%m/%d %H:%M", time.localtime()) 

v_no = 0
for s in soup.find(class_="part_list_2").find_all('h3'):
    v_no +=1
    v_date, v_time = s.find(class_="date").text.split(" ")
    if today.split(" ")[0] != v_date:
        break
    v_subject = s.find_all('a')[-1].text
    print("[%03d] %s %s - %s" % (int(v_no), v_date, v_time, v_subject))
```

## 2. 取出現在時間兩小時內的新聞

```python
v_no = 0
for s in soup.find(class_="part_list_2").find_all('h3'):
    v_no +=1
    v_date, v_time = s.find(class_="date").text.split(" ")
    if int(today.split(" ")[1].split(":")[0]) - int(v_time.split(":")[0]) > 2:
        break
    v_subject = s.find_all('a')[-1].text
    print("[%03d] %s %s - %s" % (int(v_no), v_date, v_time, v_subject))
```

```python
# 關閉瀏覽器
browser.quit();
```

```python
url = 'https://www.ettoday.net/news/news-list-{}-{}-{}-0.htm'.format("2020", "01", "30")
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(url)
```

```python
# 每個兩秒鐘自動往下滑
for r in range(50):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight),1000;")
```

```python
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
last_no, last_date, last_time, last_subject = get_last_update(soup)
print(last_no, last_date, last_time)
```

```python
v_no = 0
for s in soup.find(class_="part_list_2").find_all('h3'):
    v_no +=1
    v_date, v_time = s.find(class_="date").text.split(" ")
    v_hour = int(v_time.split(":")[0])
    if v_hour > 15 and v_hour <= 17:
        v_subject = s.find_all('a')[-1].text
        print("[%03d] %s %s - %s" % (int(v_no), v_date, v_time, v_subject))
```
