### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
len(soup.find(class_="part_list_2").find_all('h3'))
```

```python
for d in soup.find(class_="part_list_2").find('h3'):
    print(d)

len(soup.find(class_="part_list_2").find('h3'))
```

```python
i = 0
for d in soup.find(class_="part_list_2").find_all('h3'):
    i += 1    
    if i > 3:
        break
    print(d.find('span').text)
```

```python
i = 0
for d in soup.find(class_="part_list_2").find_all('h3'):
    i += 1
    if i > 3:
        break
    print(d.find('em').text)
```

```python
i = 0
for d in soup.find(class_="part_list_2").find_all('h3'):
    i += 1
    if i > 3:
        break
    print(d.find('a').text)
```

```python
import pandas as pd
columns = ['date', 'tag', 'text']
table = []
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    table.append([d.find('span').text, d.find('em').text, d.find('a').text])
```

```python
print(table[0:3])
```

```python
df = pd.DataFrame(table, columns=columns)
```

```python
df.sort_values('tag')
```
